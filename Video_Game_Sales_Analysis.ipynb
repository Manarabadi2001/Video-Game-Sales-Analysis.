```python
import pandas as pd
```

```python
# 1. تحميل البيانات واكتشافها
df = pd.read_csv('vgsales.csv')
```

```python
#أطهار أول 5 صفوف من البيانات
df.head()
```

```python
#3. تحليل المبيعات وتصوير البيانات
import matplotlib.pyplot as plt

# 1. تجميع البيانات حسب النوع، وجمع المبيعات، ثم ترتيبها من الأكبر للأصغر
top_genres = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

# 2. عرض الأرقام
print(top_genres)

# 3. رسم النتيجة لتكون واضحة للمدير
top_genres.plot(kind='bar', color='green', title='إجمالي المبيعات حسب نوع اللعبة')
plt.show()
```

```python
# 1. الفلترة: نختار فقط الصفوف التي نوعها Action
action_df = df[df['Genre'] == 'Action']

# 2. التجميع: نجمع مبيعات ألعاب الأكشن حسب المنصة
platform_sales = action_df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10) # نأخذ أعلى 10 فقط

# 3. الرسم
print(platform_sales)
platform_sales.plot(kind='bar', color='blue', title='Top Platforms for Action Games')
plt.show()
```

```python
# 1. نفلتر البيانات لنأخذ ألعاب الأكشن فقط
action_df = df[df['Genre'] == 'Action']

# 2. نجمع المبيعات حسب "السنة"
sales_by_year = action_df.groupby('Year')['Global_Sales'].sum()

# 3. الرسم الخطي (Line Chart)
print(sales_by_year.tail(10)) # طباعة آخر 10 سنوات كأرقام
sales_by_year.plot(kind='line', grid=True, title='Action Games Sales Trend Over Years')
plt.show()
```
